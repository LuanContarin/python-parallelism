Este programa tem como objetivo calcular a média de altura (em cm) para cada time da NBA, agrupados por temporada (2022-23, etc...), utilizando paralelismo.

Para referência, o algoritmo utilizará um arquivo de dados disponibilizado no Kaggle que pode ser acessado através [deste link](https://www.kaggle.com/datasets/justinas/nba-players-data)

Importação do arquivo de dados

```python
import pandas as pd
import os

dados_nba = pd.read_csv('all_seasons.csv')
```

Método responsável por calcular a média de altura dos times para cada temporada

```python
def calcular_media_altura(ano, temporada):
  teams = temporada.groupby('team_abbreviation')
  media_altura = teams['player_height'].mean()

  return (ano, media_altura)
```

Uso da biblioteca **joblib** para implementar o paralelismo e chamar a função "calcular_media_altura" para todas as temporadas

```python
from joblib import Parallel, delayed
import time

temporadas = dados_nba.groupby('season')

tempo_inicial = time.time()
results = Parallel(n_jobs=5)(delayed(calcular_media_altura)(ano, temporada) for ano, temporada in temporadas)

for ano, result in results:
  print(f'Temporada {ano}')
  for team, avg in result.items():
    print(f'\t{team} -> {avg:.2f}cm')

print(f'Elapsed time: {time.time() - tempo_inicial}')
```
